```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# İlk sayfa numarası
page_number = 1
# Dolaşılacak maksimum sayfa sayısı
max_pages = 8

# Liste oluştur
liste3 = []

while page_number <= max_pages:
    url = f"https://easycep.com/kategori/apple-2?page={page_number}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"}

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, "lxml")

    # Ürünlerin bulunduğu bölümü belirle
    st3 = soup.find_all("div", attrs={"class": "product__typeOne"})

    # Her bir ürün için
    for detaylar in st3:
        link_sonu = detaylar.a.get("href")
        link_basi = "https://easycep.com/"
        link = link_basi + link_sonu
        print(link)

        r1 = requests.get(link, headers=headers)
        soup1 = BeautifulSoup(r1.content, "lxml")

        fiyat = soup1.find("div", attrs={"class": "productDetail__price--text"}).text.strip()
        print(fiyat)

        urun_adi = soup1.find("h1", attrs={"class": "productDetail__title"}).text.strip()
        print(urun_adi)

        liste3.append([urun_adi, fiyat, link])

    page_number += 1
    print("Sayfa işlemi tamamlandı.")

# Liste kontrolü
print(liste3)
```

```python
df = pd.DataFrame(liste3)

df.columns = ["URUN_ADI","FIYAT","LINK"]
#df
df.to_excel("easy-iphone-web-scraping.xlsx")
```
