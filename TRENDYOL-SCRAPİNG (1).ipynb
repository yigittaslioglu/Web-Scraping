```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

excel = []

# İlk sayfa numarası
page_number = 1
# Dolaşılacak maksimum sayfa sayısı
max_pages = 77

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
}

while page_number <= max_pages:
    url = f"https://www.trendyol.com/yenilenmis-cep-telefonu-x-c109461?pi={page_number}"
    html = requests.get(url, headers=headers).content
    soup = BeautifulSoup(html, "html.parser")

    liste = soup.find_all("div", {"class": "p-card-chldrn-cntnr card-border"})

    for div in liste:
        product_name = div.h3.text.strip()
        price_span = div.find("div", {"class": "prc-box-dscntd"})
        price = price_span.text.strip() if price_span else "Fiyat bulunamadı"

        excel.append([product_name, price])
        print(product_name, ":", price)

    # Sayfa numarasını artır
    page_number += 1

print("Tüm sayfalar dolaşıldı.")

# İşlenen verileri Excel dosyasına aktarabilirsiniz
# Örnek olarak pandas kullanarak bir DataFrame oluşturabilir ve bu DataFrame'i Excel dosyasına yazabilirsiniz
# Örnek olarak:
#df = pd.DataFrame(excel, columns=['Ürün Adı', 'Fiyat'])
#df.to_excel('teknosa_urunler.xlsx', index=False)
```

```python
df = pd.DataFrame(excel, columns=['Ürün Adı', 'Fiyat'])
df.to_excel('trendyol-tum-cihazlar.xlsx', index=False)
```
