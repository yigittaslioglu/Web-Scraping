```python
def mainde_dolas(maingroup_label, selfservis):
    
    selfservis = notseamles[(notseamles["contact_reason_MainGroup"] == maingroup_label) & (notseamles["isSelfService"] == selfservis)]
    
    """
    burada maingroup labelları dolaşılıp self servisi 1 olanlar ve 0 olanların oranları bulunmaya çalışılacaktır
    """
    return selfservis
```

```python
def mainde_dolas(maingroup_label, selfservis):
    # Belirli bir MainGroup ve selfservis değerine sahip verileri filtreler
    filtered_df = notseamles[(notseamles["contact_reason_MainGroup"] == maingroup_label) & 
                             (notseamles["isSelfService"] == selfservis)]
    return filtered_df

# contact_reason_MainGroup'taki unique değerler üzerinde döngü
for label in notseamles["contact_reason_MainGroup"].unique():
    # selfservis değeri 1 olanlar için
    df_selfservis_1 = mainde_dolas(label, 1)
    # selfservis değeri 0 olanlar için
    df_selfservis_0 = mainde_dolas(label, 0)

    # Her bir grup için adet ve yüzde hesaplama
    total_count = len(notseamles[notseamles["contact_reason_MainGroup"] == label])
    count_selfservis_1 = len(df_selfservis_1)
    count_selfservis_0 = len(df_selfservis_0)

    # Yüzdelik oranları hesaplama
    percentage_selfservis_1 = (count_selfservis_1 / total_count) * 100
    percentage_selfservis_0 = (count_selfservis_0 / total_count) * 100

    # Sonuçları yazdırma
    print(f"Label: {label}")
    print(f"  Toplam Adet: {total_count}")
    print(f"  SelfServis 1: {count_selfservis_1} ({percentage_selfservis_1:.2f}%)")
    print(f"  SelfServis 0: {count_selfservis_0} ({percentage_selfservis_0:.2f}%)")
    print("-" * 50)
```

```python
def csrlevel_3_dolas(maingroup_label, selfservis):
    # Belirli bir MainGroup ve selfservis değerine sahip verileri filtreler
    filtered_df = notseamles[(notseamles["contact_reason_CSRLevel3"] == maingroup_label) & 
                             (notseamles["isSelfService"] == selfservis)]
    return filtered_df

# contact_reason_MainGroup'taki unique değerler üzerinde döngü
for label in notseamles["contact_reason_CSRLevel3"].unique():
    # selfservis değeri 1 olanlar için
    df_selfservis_1 = csrlevel_3_dolas(label, 1)
    # selfservis değeri 0 olanlar için
    df_selfservis_0 = csrlevel_3_dolas(label, 0)

    # Her bir grup için adet ve yüzde hesaplama
    total_count = len(notseamles[notseamles["contact_reason_CSRLevel3"] == label])
    count_selfservis_1 = len(df_selfservis_1)
    count_selfservis_0 = len(df_selfservis_0)

    # Yüzdelik oranları hesaplama
    percentage_selfservis_1 = (count_selfservis_1 / total_count) * 100
    percentage_selfservis_0 = (count_selfservis_0 / total_count) * 100

    # Sonuçları yazdırma
    print(f"Label: {label}")
    print(f"  Toplam Adet: {total_count}")
    print(f"  SelfServis 1: {count_selfservis_1} ({percentage_selfservis_1:.2f}%)")
    print(f"  SelfServis 0: {count_selfservis_0} ({percentage_selfservis_0:.2f}%)")
    print("-" * 50)
```

```python
# Top 5 alanı buluyoruz
top_5_areas = self_ser_cat_3.groupby('contact_reason_CSRLevel3').size().nlargest(5).index

# Her alan için NPS-Q-Score frekansları ve NPS skorunu hesaplıyoruz
for area in top_5_areas:
    print(f"Order Area: {area}")
    
    # İlgili alan için order_kitchen frekanslarını bul
    kitchen_counts = self_ser_cat_3[self_ser_cat_3['contact_reason_CSRLevel3'] == area]['NPS-Q-Score'].value_counts()
    
    # NPS skoru için verileri ayırıyoruz
    data = self_ser_cat_3[self_ser_cat_3['contact_reason_CSRLevel3'] == area]['NPS-Q-Score']
    total_responses = len(data)
    promoters = len(data[(data >= 9) & (data <= 10)])  # 9-10 puan alanlar
    detractors = len(data[(data >= 1) & (data <= 6)])  # 1-6 puan alanlar
    
    # NPS skorunu hesaplıyoruz
    if total_responses > 0:
        nps_score = (promoters - detractors) / total_responses * 100
    else:
        nps_score = 0

    # En çok kullanılan kitchen değerlerini yazdırıyoruz
    print(kitchen_counts.head(20))  # İlk 20 değeri göstermek için .head() kullanıyoruz
    print(f"NPS-SCORE: {nps_score:.2f}%\n")
```
